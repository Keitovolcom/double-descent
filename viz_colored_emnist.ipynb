```python
# 普通に可視化

import numpy as np
import matplotlib.pyplot as plt

# データセットの読み込み
x_train = np.load('./data/distribution_colored_EMNIST_Seed42_Var10000_Corr0.5/x_train_colored.npy')
y_train_colors = np.load('./data/distribution_colored_EMNIST_Seed42_Var10000_Corr0.5/y_train_colors.npy')
y_train_digits = np.load('./data/distribution_colored_EMNIST_Seed42_Var10000_Corr0.5/y_train_digits.npy')

x_test = np.load('data/distribution_colored_EMNIST_Seed42_Var10000_Corr0.5/x_test_colored.npy')
y_test_colors = np.load('data/distribution_colored_EMNIST_Seed42_Var10000_Corr0.5/y_test_colors.npy')
y_test_digits = np.load('data/distribution_colored_EMNIST_Seed42_Var10000_Corr0.5/y_test_digits.npy')

# 各数字と色の画像例を一枚ずつ表示する
def plot_examples(x, y_digits, y_colors, dataset_name='Train'):
    plt.figure(figsize=(15, 15))
    for digit in range(10):
        for color in range(10):
            index = np.where((y_digits == digit) & (y_colors == color))[0][0]
            plt.subplot(10, 10, digit * 10 + color + 1)
            # 画像を正しく表示するために転置
            plt.imshow(np.transpose(x[index], (1, 0, 2)))
            plt.axis('off')
            plt.title(f'{digit}-{color}', fontsize=8, pad=3)
    plt.suptitle(f'{dataset_name} Dataset Examples', fontsize=16)
    plt.subplots_adjust(wspace=0.5, hspace=0.5)
    plt.show()

# トレーニングセットの例を表示
plot_examples(x_train, y_train_digits, y_train_colors, 'Train')

# テストセットの例を表示
plot_examples(x_test, y_test_digits, y_test_colors, 'Test')
```

```python
# 正規化して可視化
import numpy as np
import matplotlib.pyplot as plt

# データセットの読み込み
path = 'data/distribution_colored_EMNIST_Seed42_Var10000_Corr0.5'
x_train = np.load(f'{path}/x_train_colored.npy')
y_train_colors = np.load(f'{path}/y_train_colors.npy')
y_train_digits = np.load(f'{path}/y_train_digits.npy')

x_test = np.load(f'{path}/x_test_colored.npy')
y_test_colors = np.load(f'{path}/y_test_colors.npy')
y_test_digits = np.load(f'{path}/y_test_digits.npy')

# 正規化
x_train_normalized = x_train / 255.0
x_test_normalized = x_test / 255.0

# 各数字と色の画像例を一枚ずつ表示する関数
def plot_examples(x, y_digits, y_colors, dataset_name='Train'):
    plt.figure(figsize=(15, 15))
    for digit in range(10):
        for color in range(10):
            index = np.where((y_digits == digit) & (y_colors == color))[0][0]
            plt.subplot(10, 10, digit * 10 + color + 1)
            # 画像を正しく表示するために転置
            plt.imshow(np.transpose(x[index], (1, 0, 2)))
            plt.axis('off')
            plt.title(f'{digit}-{color}', fontsize=8, pad=3)
    plt.suptitle(f'{dataset_name} Dataset Examples', fontsize=16)
    plt.subplots_adjust(wspace=0.5, hspace=0.5)
    plt.show()

# トレーニングセットの例を表示
plot_examples(x_train_normalized, y_train_digits, y_train_colors, 'Train')

# テストセットの例を表示
plot_examples(x_test_normalized, y_test_digits, y_test_colors, 'Test')
```
