```python
import pandas as pd
import matplotlib.pyplot as plt
import glob
import os

# カラーマップの定義
colors = {
    'combined': 'blue',
    'digit': 'green',
    'color': 'red'
}

# ディレクトリ内の全てのlog.csvを検索
root_path = "/workspace/csv/combine/"
directories = glob.glob(root_path + "*/")

# 初期化
all_epochs = []

# データ収集
for directory in directories:
    log_files = glob.glob(os.path.join(directory, "log.csv"))
    if not log_files:
        print(f"No log.csv found in {directory}")
        continue

    for log_file in log_files:
        df = pd.read_csv(log_file)
        all_epochs.extend(df['epoch'])

# エポックの範囲を計算
x_min, x_max = 1, 1000
acc_min, acc_max = 0, 1

# データの読み込みと可視化を各ディレクトリで行う関数
def process_and_visualize_log_files(directory):
    log_files = glob.glob(os.path.join(directory, "log.csv"))
    if not log_files:
        print(f"No log.csv found in {directory}")
        return

    for log_file in log_files:
        df = pd.read_csv(log_file)

        epochs = df['epoch']
        train_loss = df[' train_loss']
        test_loss = df[' test_loss']
        train_combined_acc = df[' train_combined_acc']
        test_combined_acc = df[' test_combined_acc']
        train_digit_acc = df[' train_digit_acc']
        test_digit_acc = df[' test_digit_acc']
        train_color_acc = df[' train_color_acc']
        test_color_acc = df[' test_color_acc']

        # 損失の範囲を計算
        loss_min, loss_max = min(train_loss.min(), test_loss.min()), max(train_loss.max(), test_loss.max())

        # 訓練正解率のグラフ
        plt.figure(figsize=(12, 12))
        plt.plot(epochs, train_combined_acc, label='Train Combined Accuracy', color=colors['combined'])
        plt.plot(epochs, train_digit_acc, label='Train Digit Accuracy', color=colors['digit'])
        plt.plot(epochs, train_color_acc, label='Train Color Accuracy', color=colors['color'])
        plt.xscale('log')
        plt.xlim(x_min, x_max)
        plt.ylim(acc_min, acc_max)
        plt.xlabel('Epoch')
        plt.ylabel('Accuracy')
        plt.legend()
        plt.title('Training Accuracies')
        plt.grid(True)
        output_path = os.path.join(directory, 'train_accuracies.png')
        plt.savefig(output_path)
        plt.close()

        # テスト正解率のグラフ
        plt.figure(figsize=(12, 12))
        plt.plot(epochs, test_combined_acc, label='Test Combined Accuracy', color=colors['combined'])
        plt.plot(epochs, test_digit_acc, label='Test Digit Accuracy', color=colors['digit'])
        plt.plot(epochs, test_color_acc, label='Test Color Accuracy', color=colors['color'])
        plt.xscale('log')
        plt.xlim(x_min, x_max)
        plt.ylim(acc_min, acc_max)
        plt.xlabel('Epoch')
        plt.ylabel('Accuracy')
        plt.legend()
        plt.title('Test Accuracies')
        plt.grid(True)
        output_path = os.path.join(directory, 'test_accuracies.png')
        plt.savefig(output_path)
        plt.close()

        # 訓練損失のグラフ
        plt.figure(figsize=(12, 12))
        plt.plot(epochs, train_loss, label='Train Loss', color='blue')
        plt.xscale('log')
        plt.xlim(x_min, x_max)
        plt.ylim(loss_min, loss_max)
        plt.xlabel('Epoch')
        plt.ylabel('Loss')
        plt.legend()
        plt.title('Training Loss')
        plt.grid(True)
        output_path = os.path.join(directory, 'train_loss.png')
        plt.savefig(output_path)
        plt.close()

        # テスト損失のグラフ
        plt.figure(figsize=(12, 12))
        plt.plot(epochs, test_loss, label='Test Loss', color='blue')
        plt.xscale('log')
        plt.xlim(x_min, x_max)
        plt.ylim(loss_min, loss_max)
        plt.xlabel('Epoch')
        plt.ylabel('Loss')
        plt.legend()
        plt.title('Test Loss')
        plt.grid(True)
        output_path = os.path.join(directory, 'test_loss.png')
        plt.savefig(output_path)
        plt.close()

        print(f"Visualizations saved in {directory}")

# 指定されたパスのすべてのサブディレクトリを処理
for directory in directories:
    process_and_visualize_log_files(directory)
```

```python
import warnings
import pandas as pd
import matplotlib.pyplot as plt
import glob
import os

warnings.filterwarnings('ignore')

# カラーマップの定義
colors = {
    'combined': 'blue',
    'digit': 'green',
    'color': 'red'
}

# ディレクトリ内の全てのlog.csvを検索
root_path = "/workspace/csv/combine/"
directories = glob.glob(root_path + "*/")

# Varianceごとにディレクトリを分類
variance_dirs = {100: [], 1000: [], 10000: [], 100000: []}
for directory in directories:
    if "Variance100" in directory:
        variance_dirs[100].append(directory)
    elif "Variance1000" in directory:
        variance_dirs[1000].append(directory)
    elif "Variance10000" in directory:
        variance_dirs[10000].append(directory)
    elif "Variance100000" in directory:
        variance_dirs[100000].append(directory)

# 横並びのグラフを作成し保存する関数
def create_combined_plots(variance, directories):
    fig, axs = plt.subplots(4, 4, figsize=(20, 20))
    fig.suptitle(f'Variance: {variance}', fontsize=16)

    for i, directory in enumerate(directories):
        log_file = os.path.join(directory, "log.csv")
        if not os.path.exists(log_file):
            continue

        df = pd.read_csv(log_file)

        epochs = df['epoch']
        train_loss = df['train_loss']
        test_loss = df['test_loss']
        train_combined_acc = df['train_combined_acc']
        test_combined_acc = df['test_combined_acc']
        train_digit_acc = df['train_digit_acc']
        test_digit_acc = df['test_digit_acc']
        train_color_acc = df['train_color_acc']
        test_color_acc = df['test_color_acc']

        loss_min, loss_max = min(train_loss.min(), test_loss.min()), max(train_loss.max(), test_loss.max())
        
        row = i // 4
        col = i % 4

        # 訓練正解率のグラフ
        axs[row, col].plot(epochs, train_combined_acc, label='Train Combined Accuracy', color=colors['combined'])
        axs[row, col].plot(epochs, train_digit_acc, label='Train Digit Accuracy', color=colors['digit'])
        axs[row, col].plot(epochs, train_color_acc, label='Train Color Accuracy', color=colors['color'])
        axs[row, col].set_xscale('log')
        axs[row, col].set_xlim(1, 1000)
        axs[row, col].set_ylim(0, 1)
        axs[row, col].set_xlabel('Epoch')
        axs[row, col].set_ylabel('Accuracy')
        axs[row, col].legend()
        axs[row, col].set_title('Training Accuracies')
        axs[row, col].grid(True)

        # テスト正解率のグラフ
        axs[row + 1, col].plot(epochs, test_combined_acc, label='Test Combined Accuracy', color=colors['combined'])
        axs[row + 1, col].plot(epochs, test_digit_acc, label='Test Digit Accuracy', color=colors['digit'])
        axs[row + 1, col].plot(epochs, test_color_acc, label='Test Color Accuracy', color=colors['color'])
        axs[row + 1, col].set_xscale('log')
        axs[row + 1, col].set_xlim(1, 1000)
        axs[row + 1, col].set_ylim(0, 1)
        axs[row + 1, col].set_xlabel('Epoch')
        axs[row + 1, col].set_ylabel('Accuracy')
        axs[row + 1, col].legend()
        axs[row + 1, col].set_title('Test Accuracies')
        axs[row + 1, col].grid(True)

        # 訓練損失のグラフ
        axs[row + 2, col].plot(epochs, train_loss, label='Train Loss', color='blue')
        axs[row + 2, col].set_xscale('log')
        axs[row + 2, col].set_xlim(1, 1000)
        axs[row + 2, col].set_ylim(loss_min, loss_max)
        axs[row + 2, col].set_xlabel('Epoch')
        axs[row + 2, col].set_ylabel('Loss')
        axs[row + 2, col].legend()
        axs[row + 2, col].set_title('Training Loss')
        axs[row + 2, col].grid(True)

        # テスト損失のグラフ
        axs[row + 3, col].plot(epochs, test_loss, label='Test Loss', color='blue')
        axs[row + 3, col].set_xscale('log')
        axs[row + 3, col].set_xlim(1, 1000)
        axs[row + 3, col].set_ylim(loss_min, loss_max)
        axs[row + 3, col].set_xlabel('Epoch')
        axs[row + 3, col].set_ylabel('Loss')
        axs[row + 3, col].legend()
        axs[row + 3, col].set_title('Test Loss')
        axs[row + 3, col].grid(True)

    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    combined_dir = os.path.join(root_path, 'combined_plots')
    os.makedirs(combined_dir, exist_ok=True)
    output_path = os.path.join(combined_dir, f'combined_plots_variance_{variance}.png')
    plt.savefig(output_path)
    plt.close()

# 各Varianceごとにプロットを作成
for variance, dirs in variance_dirs.items():
    create_combined_plots(variance, dirs)
```

```python
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages

def create_legend_plot():
    # ラベルと色の関係
    labels = [
        'Train Combined Accuracy', 'Test Combined Accuracy', 
        'Train Digit Accuracy', 'Test Digit Accuracy', 
        'Train Color Accuracy', 'Test Color Accuracy', 
        'Train Loss', 'Test Loss'
    ]
    colors = [
        'blue', 'blue', 
        'green', 'green', 
        'red', 'red', 
        'blue', 'blue'
    ]
    linestyles = [
        'solid', 'dashed', 
        'solid', 'dashed', 
        'solid', 'dashed', 
        'solid', 'dashed'
    ]

    fig, ax = plt.subplots(figsize=(8, 6))

    # ダミーデータをプロットして凡例を作成
    for label, color, linestyle in zip(labels, colors, linestyles):
        ax.plot([], [], label=label, color=color, linestyle=linestyle)

    # 凡例を表示
    ax.legend(loc='center', fontsize='medium')
    ax.axis('off')  # 軸をオフにして空の図にする

    plt.tight_layout()
    
    return fig

# PDFに保存する関数
def save_legend_to_pdf(filename):
    fig = create_legend_plot()
    with PdfPages(filename) as pdf:
        pdf.savefig(fig, bbox_inches='tight')  # 余白をなくすために bbox_inches='tight' を追加

# プログラムの実行とPDF保存
save_legend_to_pdf('legend_plot.pdf')

# 画面に表示
plt.show()
```
